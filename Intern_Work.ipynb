```python
import pandas as pd
import matplotlib.pyplot as plt

annual_sales = df.groupby(['year', 'Store'])['Weekly_Sales'].sum().unstack()


for year in annual_sales.index:
    plt.figure(figsize=(10, 10))
    plt.pie(annual_sales.loc[year], labels=annual_sales.columns,
            autopct=lambda p: '{:.1f}%'.format(p) if p > 0 else '',
            startangle=140, colors=plt.cm.Paired.colors, textprops={'fontsize': 10})
    plt.title(f'Annual Sales Distribution by Store - {year}', fontsize=14)
    plt.show()
```
